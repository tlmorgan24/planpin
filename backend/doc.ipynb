```python
from docx import Document
from docx.shared import Inches
from docx.enum.table import WD_TABLE_ALIGNMENT
from docx.enum.text import WD_ALIGN_PARAGRAPH
import matplotlib.pyplot as plt
import os

# Step 1: Create a sample chart using matplotlib
chart_filename = "sample_chart.png"
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.figure(figsize=(5, 3))
plt.plot(x, y, marker='o')
plt.title("Sample Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.tight_layout()
plt.savefig(chart_filename)
plt.close()

# Step 2: Create a Word document
doc = Document("C:\\Users\\MOR112203\\Downloads\\template.docx")

# Title & subtitle:
doc.add_heading("Monthly Report", level=0)
doc.add_paragraph("This is an automatically generated report containing a sample table and a chart.", style="Subtitle")

doc.add_heading("Table", level=1)
doc.add_heading("Subheading", level=2)
doc.add_paragraph("List test:")
doc.add_paragraph("List 1", style='List Paragraph')
doc.add_paragraph("List 2", style='List Paragraph')
doc.add_paragraph("List 3", style='List Paragraph')

# Add the chart image
doc.add_paragraph("Defect test:")

# Add a table with sample data
table = doc.add_table(rows=6, cols=2)
table.style = 'Table Grid'
table.alignment = WD_TABLE_ALIGNMENT.CENTER

header_1 = table.cell(0, 0)
header_2 = table.cell(0, 1)
header = header_1.merge(header_2)
header.text = "Defect reference: ..."

table.cell(1,0).text = "Category"
table.cell(1,1).text = "Cracks"

table.cell(2,0).text = "Description"
table.cell(2,1).text = "2mm crack. I am doing a test for quite a long description to se sdgs skdfh sdkjfsdhyhwesd mhnweicss bui "

table.cell(3,0).text = "Severity"
table.cell(3,1).text = "5"

for i in range(1,4):
    left_cell = table.cell(i, 0)
    right_cell = table.cell(i, 0)
    left_cell.width = Inches(1.2)

image_cell1 = table.cell(4, 0)
image_cell2 = table.cell(4, 1)
image_cell = image_cell1.merge(image_cell2)
paragraph = image_cell.paragraphs[0]
paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER # for centred image
run = paragraph.add_run()
run.add_picture("sample_chart.png", width=Inches(5))

caption_1 = table.cell(5, 0)
caption_2 = table.cell(5, 1)
caption = caption_1.merge(caption_2)
paragraph = caption.paragraphs[0]
paragraph.style = 'Caption'
run = paragraph.add_run("This is a caption")


doc.add_picture(chart_filename, width=Inches(5.5))
doc.add_picture(chart_filename, width=Inches(5.5))
doc.add_picture(chart_filename, width=Inches(5.5))

# Save the document
doc.save("C:\\Users\\MOR112203\\Downloads\\test-doc.docx")

```
